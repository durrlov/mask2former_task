```python
# Import

import json
import cv2
import numpy as np
import os
import rasterio
from shapely.geometry import Polygon
from PIL import Image
import math
from tqdm import tqdm

# path
RAW_DATA_DIR = "/content/drive/MyDrive/mask2former_task/tree_data_for_instance_segmentation"

PREPROCESSED_DIR = "/content/drive/MyDrive/mask2former_task/preprocessed_data"
TRACKING_JSON_PATH = "/content/drive/MyDrive/mask2former_task/TRACKING.json"

# Preprocessing parameters
PATCH_SIZE = 1024
STRIDE = 512
KEEP_THRESHOLD = 0.50
RESIZE_FACTOR = 0.60


# Main Processing Function


def preprocess_data(image_files, raw_data_dir, preprocessed_dir, tracking_json_path, is_train=True):
    """
    Preprocess tree segmentation data:
    1. Resize images to reduce memory usage
    2. Pad to multiples of patch size
    3. Extract patches with stride
    4. Filter objects based on overlap threshold
    5. Create instance masks for each patch

    Args:
        image_files: List of .tif filenames to process
        raw_data_dir: Directory containing .tif images and .json annotations
        preprocessed_dir: Output directory for processed patches
        tracking_json_path: Path to tracking JSON (prevents reprocessing)
        is_train: If True, save to train folders; else test folders
    """

    # Load or initialize tracking
    try:
        with open(tracking_json_path, 'r') as fp:
            tracking = json.load(fp)
        preprocessed_files = tracking["preprocessed"]
        latest_start_id = tracking["latest_start_id"] + 1000
    except:
        tracking = {
            "preprocessed": [],
            "latest_start_id": 1000,
            "mapping": {
                "id_to_filename": [],
                "filename_to_id": []
            }
        }
        preprocessed_files = []
        latest_start_id = 1000

    # Create output folders
    if is_train:
        out_dir_png = os.path.join(preprocessed_dir, "train_PNG")
        out_dir_mask = os.path.join(preprocessed_dir, "train_MASK")
        out_dir_bbox = os.path.join(preprocessed_dir, "train_BBOX")
    else:
        out_dir_png = os.path.join(preprocessed_dir, "val_PNG")
        out_dir_mask = os.path.join(preprocessed_dir, "val_MASK")
        out_dir_bbox = os.path.join(preprocessed_dir, "val_BBOX")

    os.makedirs(out_dir_png, exist_ok=True)
    os.makedirs(out_dir_mask, exist_ok=True)
    os.makedirs(out_dir_bbox, exist_ok=True)

    print(f"\n{'='*70}")
    print(f"Processing {'TRAINING' if is_train else 'VALIDATION'} data")
    print(f"{'='*70}")
    print(f"Output directories:")
    print(f"  - Images: {out_dir_png}")
    print(f"  - Masks:  {out_dir_mask}")
    print(f"  - BBoxes: {out_dir_bbox}\n")

    total_patches = 0

    # Process each image
    for image_file in tqdm(image_files, desc=f"Processing {'train' if is_train else 'val'} files"):

        # Skip if already processed
        if image_file in preprocessed_files:
            print(f"\n‚ö† WARNING: {image_file} already processed. Skipping...")
            print(f"   To reprocess, remove from {tracking_json_path}")
            continue

        print(f"\nüìÑ Processing: {image_file}")

        tracking["preprocessed"].append(image_file)
        tracking["latest_start_id"] = latest_start_id
        tracking["mapping"]["id_to_filename"].append({latest_start_id: image_file})
        tracking["mapping"]["filename_to_id"].append({image_file: latest_start_id})

        filename = image_file.replace('.tif', '')

        ann_path = os.path.join(raw_data_dir, f"{filename}.json")
        with open(ann_path) as f:
            data = json.load(f)

        img_path = os.path.join(raw_data_dir, image_file)
        with rasterio.open(img_path) as src:
            image = np.moveaxis(src.read(), 0, -1)

            # Take only RGB channels if more exist
            if image.shape[-1] > 3:
                image = image[:, :, :3]

            H, W, _ = image.shape
            print(f"   Original size: {W}x{H}")

            new_H = int(H * RESIZE_FACTOR)
            new_W = int(W * RESIZE_FACTOR)
            image_resized = cv2.resize(image, (new_W, new_H), interpolation=cv2.INTER_LINEAR)
            print(f"   Resized to: {new_W}x{new_H}")

            pad_h = (math.ceil(new_H / PATCH_SIZE) * PATCH_SIZE) - new_H
            pad_w = (math.ceil(new_W / PATCH_SIZE) * PATCH_SIZE) - new_W
            image_padded = np.pad(
                image_resized,
                ((0, pad_h), (0, pad_w), (0, 0)),
                mode="reflect"
            )
            H_pad, W_pad, _ = image_padded.shape
            print(f"   Padded to: {W_pad}x{H_pad}")

        shapes_scaled = []
        for shape in data["shapes"]:
            pts = np.array(shape["points"], dtype=float)
            pts *= RESIZE_FACTOR
            shapes_scaled.append({**shape, "points": pts.tolist()})

        print(f"   Original annotations: {len(shapes_scaled)} trees")

        # Extract patches with sliding window
        patch_id = latest_start_id
        patches_saved = 0

        for y in range(0, H_pad - PATCH_SIZE + 1, STRIDE):
            for x in range(0, W_pad - PATCH_SIZE + 1, STRIDE):
                patch = image_padded[y:y+PATCH_SIZE, x:x+PATCH_SIZE]

                # Find which objects overlap with this patch
                patch_shapes = []
                patch_box = Polygon([
                    (x, y),
                    (x + PATCH_SIZE, y),
                    (x + PATCH_SIZE, y + PATCH_SIZE),
                    (x, y + PATCH_SIZE)
                ])

                for shape in shapes_scaled:
                    try:
                        pts = np.array(shape["points"])
                        poly = Polygon(pts)

                        # Calculate intersection
                        inter = poly.intersection(patch_box)

                        # Keep only if significant overlap
                        if not inter.is_empty and inter.area / poly.area >= KEEP_THRESHOLD:
                            if inter.geom_type == "Polygon":
                                polygons = [inter]
                            elif inter.geom_type == "MultiPolygon":
                                polygons = list(inter.geoms)
                            else:
                                continue

                            # Convert to patch-local coordinates
                            for sub_poly in polygons:
                                inter_coords = np.array(sub_poly.exterior.coords)
                                shifted = [
                                    [float(px - x), float(py - y)]
                                    for px, py in inter_coords
                                ]
                                patch_shapes.append({
                                    "label": shape["label"],
                                    "points": shifted,
                                    "group_id": None,
                                    "shape_type": "polygon",
                                    "flags": {}
                                })
                    except Exception as e:
                        print(f"      Invalid geometry, skipping: {e}")
                        pass

                # Skip patches with no annotations
                if len(patch_shapes) == 0:
                    continue

                # Save patch image
                png_name = f"{patch_id}.png"
                Image.fromarray(patch.astype(np.uint8)).save(
                    os.path.join(out_dir_png, png_name)
                )

                new_json = {
                    "version": data.get("version", "5.8.3"),
                    "flags": {},
                    "shapes": patch_shapes,
                    "imagePath": png_name,
                    "imageHeight": PATCH_SIZE,
                    "imageWidth": PATCH_SIZE,
                }
                with open(os.path.join(out_dir_png, f"{patch_id}.json"), "w") as jf:
                    json.dump(new_json, jf, indent=2)

                # Create instance mask
                mask = np.zeros((PATCH_SIZE, PATCH_SIZE), dtype=np.uint8)
                for inst_idx, shp in enumerate(patch_shapes, start=1):
                    pts = np.array(shp["points"], dtype=np.int32)
                    cv2.fillPoly(mask, [pts], inst_idx)

                mask_name = f"{patch_id}_mask.png"
                cv2.imwrite(os.path.join(out_dir_mask, mask_name), mask)

                # Create bounding box visualization
                preview = patch.copy().astype(np.uint8)
                for shp in patch_shapes:
                    xs = [p[0] for p in shp["points"]]
                    ys = [p[1] for p in shp["points"]]
                    box_x_min = int(min(xs))
                    box_y_min = int(min(ys))
                    box_x_max = int(max(xs))
                    box_y_max = int(max(ys))

                    cv2.rectangle(
                        preview,
                        (box_x_min, box_y_min),
                        (box_x_max, box_y_max),
                        (0, 255, 0), 2
                    )

                bbox_png = os.path.join(out_dir_bbox, f"{patch_id}_bbox.png")
                cv2.imwrite(bbox_png, cv2.cvtColor(preview, cv2.COLOR_RGB2BGR))

                patch_id += 1
                patches_saved += 1

        print(f"   ‚úì Created {patches_saved} patches")
        total_patches += patches_saved
        latest_start_id += 1000

        with open(tracking_json_path, 'w') as fp:
            json.dump(tracking, fp, indent=2)

    print(f"\n{'='*70}")
    print(f"‚úÖ Preprocessing complete!")
    print(f"   Total patches created: {total_patches}")
    print(f"{'='*70}\n")

    return total_patches


# Data Splitting and Organization


def split_and_organize_data(raw_data_dir, preprocessed_dir, tracking_json_path,
                            train_ratio=0.8, seed=42):
    """
    Split raw data into train/val sets and preprocess both.

    Args:
        raw_data_dir: Directory with raw .tif and .json files
        preprocessed_dir: Output directory
        tracking_json_path: Path to tracking JSON
        train_ratio: Fraction of data for training (default 0.8)
        seed: Random seed for reproducibility
    """

    print("\n" + "="*70)
    print("TREE INSTANCE SEGMENTATION - DATA PREPROCESSING")
    print("="*70)
    print(f"\nInput directory:  {raw_data_dir}")
    print(f"Output directory: {preprocessed_dir}")
    print(f"Tracking file:    {tracking_json_path}\n")

    all_files = [f for f in os.listdir(raw_data_dir) if f.endswith('.tif')]
    all_files.sort()

    print(f"Found {len(all_files)} image files\n")

    if len(all_files) == 0:
        print("‚ùå Error: No .tif files found!")
        return

    np.random.seed(seed)
    np.random.shuffle(all_files)

    n_train = int(len(all_files) * train_ratio)
    train_files = all_files[:n_train]
    val_files = all_files[n_train:]

    print(f"Dataset split (seed={seed}):")
    print(f"  Training:   {len(train_files)} images")
    print(f"  Validation: {len(val_files)} images\n")

    print("Training files:", train_files)
    print("Validation files:", val_files)

    split_info = {
        "train_files": train_files,
        "val_files": val_files,
        "train_ratio": train_ratio,
        "seed": seed
    }
    split_info_path = os.path.join(preprocessed_dir, "split_info.json")
    os.makedirs(preprocessed_dir, exist_ok=True)
    with open(split_info_path, 'w') as f:
        json.dump(split_info, f, indent=2)
    print(f"\n‚úì Split info saved to: {split_info_path}")

    # Process training data
    print("\n" + "="*70)
    print("STEP 1: Processing TRAINING data")
    print("="*70)
    train_patches = preprocess_data(
        train_files,
        raw_data_dir,
        preprocessed_dir,
        tracking_json_path,
        is_train=True
    )

    # Process validation data
    print("\n" + "="*70)
    print("STEP 2: Processing VALIDATION data")
    print("="*70)
    val_patches = preprocess_data(
        val_files,
        raw_data_dir,
        preprocessed_dir,
        tracking_json_path,
        is_train=False
    )

    # Final summary
    print("\n" + "="*70)
    print("PREPROCESSING SUMMARY")
    print("="*70)
    print(f"\nüìä Statistics:")
    print(f"  Training:")
    print(f"    - Images:  {len(train_files)}")
    print(f"    - Patches: {train_patches}")
    print(f"  Validation:")
    print(f"    - Images:  {len(val_files)}")
    print(f"    - Patches: {val_patches}")
    print(f"  Total patches: {train_patches + val_patches}")

    print(f"\nüìÅ Output structure:")
    print(f"  {preprocessed_dir}/")
    print(f"  ‚îú‚îÄ‚îÄ train_PNG/      ({train_patches} images)")
    print(f"  ‚îú‚îÄ‚îÄ train_MASK/     ({train_patches} masks)")
    print(f"  ‚îú‚îÄ‚îÄ train_BBOX/     ({train_patches} visualizations)")
    print(f"  ‚îú‚îÄ‚îÄ val_PNG/        ({val_patches} images)")
    print(f"  ‚îú‚îÄ‚îÄ val_MASK/       ({val_patches} masks)")
    print(f"  ‚îú‚îÄ‚îÄ val_BBOX/       ({val_patches} visualizations)")
    print(f"  ‚îú‚îÄ‚îÄ split_info.json")
    print(f"  ‚îî‚îÄ‚îÄ TRACKING.json")

    print(f"\n‚úÖ Preprocessing complete! Ready for Mask2Former training.")
    print("="*70 + "\n")



# Visualization Function


def visualize_sample(preprocessed_dir, patch_id, split="train"):
    """
    Visualize a preprocessed patch with its mask and bbox overlay.

    Args:
        preprocessed_dir: Directory containing preprocessed data
        patch_id: ID of the patch to visualize
        split: "train" or "val"
    """
    import matplotlib.pyplot as plt

    png_dir = os.path.join(preprocessed_dir, f"{split}_PNG")
    mask_dir = os.path.join(preprocessed_dir, f"{split}_MASK")
    bbox_dir = os.path.join(preprocessed_dir, f"{split}_BBOX")

    image = np.array(Image.open(os.path.join(png_dir, f"{patch_id}.png")))
    mask = cv2.imread(os.path.join(mask_dir, f"{patch_id}_mask.png"), cv2.IMREAD_GRAYSCALE)
    bbox = cv2.imread(os.path.join(bbox_dir, f"{patch_id}_bbox.png"))
    bbox = cv2.cvtColor(bbox, cv2.COLOR_BGR2RGB)

    num_instances = len(np.unique(mask)) - 1  # Exclude background

    fig, axes = plt.subplots(1, 3, figsize=(18, 6))

    axes[0].imshow(image)
    axes[0].set_title(f"Patch {patch_id}\nOriginal Image", fontsize=14)
    axes[0].axis('off')

    axes[1].imshow(mask, cmap='nipy_spectral')
    axes[1].set_title(f"Instance Mask\n{num_instances} trees", fontsize=14)
    axes[1].axis('off')

    axes[2].imshow(bbox)
    axes[2].set_title("Bounding Boxes", fontsize=14)
    axes[2].axis('off')

    plt.tight_layout()
    plt.savefig(os.path.join(preprocessed_dir, f"sample_visualization_{patch_id}.png"),
                dpi=150, bbox_inches='tight')
    plt.show()

    print(f"‚úì Visualization saved to: {preprocessed_dir}/sample_visualization_{patch_id}.png")




if __name__ == "__main__":

    # Run preprocessing with train/val split
    split_and_organize_data(
        raw_data_dir=RAW_DATA_DIR,
        preprocessed_dir=PREPROCESSED_DIR,
        tracking_json_path=TRACKING_JSON_PATH,
        train_ratio=0.8,
        seed=42
    )

    # Visualize a sample
    print("\n" + "="*70)
    print("Creating sample visualization...")
    print("="*70 + "\n")

    # Find first training patch
    train_png_dir = os.path.join(PREPROCESSED_DIR, "train_PNG")
    if os.path.exists(train_png_dir):
        png_files = [f for f in os.listdir(train_png_dir) if f.endswith('.png') and '_' not in f]
        if png_files:
            first_patch_id = png_files[0].replace('.png', '')
            visualize_sample(PREPROCESSED_DIR, first_patch_id, split="train")
```
